```python
paleo = pd.read_csv('../Data/scaled_paleo.csv', index_col = 0)
primal = pd.read_csv('../Data/scaled_primal.csv', index_col = 0)
whole30 = pd.read_csv('../Data/scaled_whole30.csv', index_col = 0)
veggie = pd.read_csv('../Data/scaled_vegetarian.csv', index_col = 0)
vegan = pd.read_csv('../Data/scaled_vegan.csv', index_col = 0)
```

```python
predictors = ['num_ingredients', 'pricePerServing', 'readyInMinutes', 'servings',
          'weightWatcherSmartPoints', 'Calories', 'Fat', 'Saturated_Fat','Carbohydrates', 
          'Sugar', 'Cholesterol', 'Sodium', 'Protein','Vitamin_K', 'Vitamin_A', 
          'Vitamin_C', 'Manganese', 'Folate', 'Fiber','Copper', 'Magnesium', 
          'Phosphorus', 'Vitamin_B6', 'Potassium','Vitamin_B1', 'Iron', 'Vitamin_B2', 
          'Vitamin_E', 'Zinc', 'Vitamin_B5','Vitamin_B3', 'Calcium', 'Selenium', 
          'num_words_instructions','num_steps_instructions']
outcome = veggie.aggregateLikes
```

```python
X = veggie[predictors]
X = sm.add_constant(X)
model = sm.OLS(outcome, X, hasconst = True).fit()
model.summary()
```
