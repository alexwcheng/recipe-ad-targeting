**Provided exponential distribution, transform dependent
variable using log.**

```python
# Create new column for log(likes)
df['log_likes'] = df.apply(lambda _: 0, axis=1)
for index, value in enumerate(df['aggregatelikes']):
    try:
        df['log_likes'][index] = math.log(df[value])
    except:
        df['log_likes'][index] = None
        
# Create new column for (likes- category)
df['likes_binary'] = df.apply(lambda _: 0, axis=1)
for index, value in enumerate(df['aggregatelikes']):
    if value > df.aggregatelikes.median():
        df['likes_binary'][index] = 1
        
# Create new column for log(likes- category)
df['log_likes_binary'] = df.apply(lambda _: 0, axis=1)
for index, value in enumerate(df['likes_binary']):
    try:
        df['log_likes_binary'][index] = math.log(df[value])
    except:
        df['log_likes_binary'][index] = None
```

**Analysis**

Likes categorized as 'high' or 'low' correlate better to the independent variables than Likes itself. It also appears that both log(Number of Likes) and log(Binary Outcome) have stronger correlations with the indepdent variables as well. However, these are strong correlates to all variables with no variation in correlation. Therefore, it seems we are somewhere between a linear regression and logistic regression in which log odds should linearly relate to predictor variables.
